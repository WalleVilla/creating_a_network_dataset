#Learning how to create a dataset with a txt file in Python

```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#read the dataset
headers = ["package-type", "time-in-seconds", "node", "osi-layer", "idk1",
           "idk2", "idk3", "tcp-agent", "bytes", "idk4", "idk5", "idk6",
           "idk7", "idk8", "idk9","idk10", "idk11", "idk12","idk13"]
df = pd.read_csv("Network_information.txt", names= headers,  sep= " ")
```

```python
#print the first 10 rows
df.head(10)
```

```python
#get rid of the "_" in the column "node"
df["node"] = df["node"].str.replace("_", "")
df
```

```python
#check de datatypes in each column
df.dtypes
```

```python
#the node column is type object, it has to be int
df["node"] = df["node"].astype("int")
df.dtypes
```

```python
#saving the important data in a new csv file
ntw_df=df[["package-type",	"time-in-seconds",	"node",	"osi-layer", "tcp-agent",	"bytes"]]
ntw_df.to_csv("network_data.csv")
ntw_df
```
